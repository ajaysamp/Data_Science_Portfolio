# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

I do not agree with the conclusion that the excess rate of readmissions is going down with increasing number of admissions. The scatter plot is not intuitive enough to nake this observation. We need to plot a regression line to have a better idea of the trend. 

```python
#get only the neccessary data in a dataframe 
x = clean_hospital_read_df[(clean_hospital_read_df['Excess Readmission Ratio'].notnull())]
df = x[['Number of Discharges','Excess Readmission Ratio']]
df.columns = ['Discharges','Excess_Readmission']
df.head()
```

```python
#scatter plot with reg line
fig = plt.subplots(figsize=(8,5))
fig = plt.scatter(np.log10(df.Discharges),df.Excess_Readmission, alpha = 0.2)
fig = plt.xlabel('Number of Discharges (log values)')
fig = plt.ylabel('Excess Readmission Ratio')
a,b = np.polyfit(np.log10(df.Discharges),df.Excess_Readmission,1)
x = np.log10(df.Discharges)
y = a * x + b
fig = plt.plot(x,y,color = 'red')
plt.show()
print ('The slope of the regression line is:',a)
```

The slope of the regressiopn line indicates that there is a marginally negative trend. However, it is hard to say if this is significant. The negative trend can be a result of extreme values and outliers. A hypothesis test is necessary to verify the negative trend. 

## Hypothesis Test

Null hypothesis: excess admission ratio is same for hospitals with discharges < 100, >1000 and all other hospitals.
Alternate hypothesis: excess admission ratio is different for hospitalls across the groups

A ANOVA analysis is needed for the comparison.

ANOVA Conditions check:
1. The hospitals are different across the groups. Hence the independence criteria is met.
2. The distribution is near normal. Need to plot the data to check.

```python
#Plots to check near normality
discharge_100 = df[df.Discharges <= 100]
discharge_1000 = df[df.Discharges >= 1000]
discharge_mid = df[(df.Discharges > 100) & (df.Discharges< 1000)]
discharge_100_sorted = sorted(discharge_100['Excess_Readmission'])
discharge_1000_sorted = sorted(discharge_1000['Excess_Readmission'])
discharge_mid_sorted = sorted(discharge_mid['Excess_Readmission'])
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
fig1 = plt.hist(discharge_100.Excess_Readmission, bins=30,normed=True,color='red')
fig1 = plt.title('Excess Readmissions for Discharge < 100')
plt.subplot(1,3,2)
fig2 = plt.hist(discharge_mid.Excess_Readmission, bins=30,normed=True,color='blue')
fig2 = plt.title('Excess Readmissions for Discharge between 100 and 1000')
plt.subplot(1,3,3)
fig3 = plt.hist(discharge_1000.Excess_Readmission, bins=30,normed=True,color='green')
fig3 = plt.title('Excess Readmissions for Discharge > 1000')
plt.tight_layout()
```

#### The plots for the three cases show a near normal distribution and the ANOVA condition is met.

```python
#Anova Test
stats.f_oneway(discharge_100_sorted, discharge_1000_sorted, discharge_mid_sorted)
```

##### The P value is extremely small and the null hypothesis can be rejected. The excess readmission ratio is significantly different for the different groups . The practical significance is that it is not just enough to look at the discharge rates (as indicated by a weak negative regression line. There are other factors affecting the excess readmission ratio and needs to be further analyzed.
